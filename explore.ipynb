```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
business_list = pd.read_csv('data/InspectionData-scraped/business_list.csv')
# business_master = pd.read_csv('data/InspectionData-scraped/business_master_withyelp.csv')
cleaned_report = pd.read_csv(
    'data/InspectionData-scraped/cleaned_report_results.csv')
cleaned_violation = pd.read_csv(
    'data/InspectionData-scraped/cleaned_violation_list.csv')
geo_master = pd.read_csv('data/InspectionData-scraped/geo_master.csv')
inspection_list = pd.read_csv(
    'data/InspectionData-scraped/inpection_list_restauranttotal.csv')
license_info = pd.read_csv('data/InspectionData-scraped/license_info.csv')
mapping = pd.read_csv('data/InspectionData-scraped/mapping.csv')
restaurant_inspection = pd.read_csv(
    'data/InspectionData-scraped/restaurant_inspections.csv')
```

#### Business List

```python
business_list['category'].value_counts()
```

```python
categories = business_list['category'].value_counts().index
counts = [x for x in business_list['category'].value_counts()]
plt.figure(figsize=(12, 6))
plt.barh(categories, counts, color='skyblue')
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('Histogram of Food Service Categories')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
